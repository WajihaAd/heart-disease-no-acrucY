# Hyper Parameter Tunning

```python
# Function for preprocessing and model evaluation
def objective(trial):
    # Initialize LabelEncoder
    label_encoder = LabelEncoder()

    # List of columns to encode
    columns_to_encode = ['smoking_status', 'work_type']

    # Apply LabelEncoder on specified columns for both X_train and X_test
    for col in columns_to_encode:
        x_train[col] = label_encoder.fit_transform(x_train[col])
        x_test[col] = label_encoder.transform(x_test[col])

    # Define the transformer pipeline with hyperparameters
    transformer = ColumnTransformer(transformers=[
        ("tnf1", OneHotEncoder(sparse_output=False, drop="first"), [0, 4, 6]),
        ("tnf3", RobustScaler(), [1, 7, 8])
    ], remainder='passthrough')

    # Set up the pipeline with feature selection and classifier
    pipeline = Pipeline([
        ('preprocessing', transformer),
        ('feature_selection', SelectKBest(score_func=f_classif, k=5))
    ])

    # Apply preprocessing and feature selection to training data
    X_train_transformed = pipeline.fit_transform(x_train, y_train)

    # Resample the data using SMOTE
    smote = SMOTE(sampling_strategy='minority', k_neighbors=5)
    X_resampled, y_resampled = smote.fit_resample(X_train_transformed, y_train)

    # Define hyperparameters for RandomForestClassifier
    n_estimators = trial.suggest_categorical('n_estimators', [10, 50, 100, 200])
    max_features = trial.suggest_categorical('max_features', ['sqrt', 'log2', None])
    criterion = trial.suggest_categorical('criterion', ['gini', 'entropy'])
    bootstrap = trial.suggest_categorical('bootstrap', [True, False])
    class_weight = trial.suggest_categorical('class_weight', [None, 'balanced'])
    ccp_alpha = trial.suggest_float('ccp_alpha', 0.0, 0.1)
    
    # Initialize and fit the model
    model = RandomForestClassifier(
        n_estimators=n_estimators,
        criterion=criterion,
        bootstrap=bootstrap,
        class_weight=class_weight,
        max_features=max_features,
        ccp_alpha=ccp_alpha
    )
    model.fit(X_resampled, y_resampled)

    # Evaluate the model on the test set
    X_test_transformed = pipeline.transform(x_test)
    y_pred = model.predict(X_test_transformed)
    accuracy = accuracy_score(y_test, y_pred)

    # Return accuracy for Optuna to maximize
    return accuracy

# Set up Optuna study
study = optuna.create_study(direction="maximize")

# Perform optimization with 10 trials
study.optimize(objective, n_trials=10)

# Print the best hyperparameters and the best accuracy found
print(f"Best trial: {study.best_trial.params}")
print(f"Best accuracy: {study.best_value:.4f}")
```

```python
best_params = study.best_trial.params

# Recreate the model with the best parameters
model = RandomForestClassifier(**best_params)

# Fit the model again with the resampled training data
model.fit(X_resampled, y_resampled)

# Now evaluate the model on the test set
accuracy = model.score(pipeline.transform(x_test), y_test)
y_pred = model.predict(pipeline.transform(x_test))

# Calculate precision, recall, F1 score, and other metrics
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
y_pred_proba = model.predict_proba(pipeline.transform(x_test))[:, 1]
roc_auc = roc_auc_score(y_test, y_pred_proba)

# Print the metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print("Confusion Matrix:")
print(conf_matrix)
print(f"ROC AUC Score: {roc_auc:.4f}")
```

```python
def evaluate_model(pipeline, model, x_train, y_train, x_test, y_test):
    # Apply SMOTE for balancing the classes
    X_train_transformed = pipeline.fit_transform(x_train, y_train)
    smote = SMOTE(sampling_strategy='auto', k_neighbors=5)
    X_resampled, y_resampled = smote.fit_resample(X_train_transformed, y_train)
    
    # Fit the model
    model.fit(X_resampled, y_resampled)
    
    # Evaluate the model
    accuracy = model.score(pipeline.transform(x_test), y_test)
    print(f"Model accuracy on test data: {accuracy:.4f}")

    # Predictions
    y_pred = model.predict(pipeline.transform(x_test))
    
    # Calculate metrics
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    conf_matrix = confusion_matrix(y_test, y_pred)
    
    # Print all metrics
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print("Confusion Matrix:")
    print(conf_matrix)
    
    # Compute ROC AUC score
    y_pred_proba = model.predict_proba(pipeline.transform(x_test))[:, 1]
    roc_auc = roc_auc_score(y_test, y_pred_proba)
    print(f"ROC AUC Score: {roc_auc:.4f}")
```

```python
# Function for preprocessing and model evaluation
def objective(trial):
    # Initialize LabelEncoder
    label_encoder = LabelEncoder()

    # List of columns to encode
    columns_to_encode = ['smoking_status', 'work_type']

    # Apply LabelEncoder on specified columns for both X_train and X_test
    for col in columns_to_encode:
        x_train[col] = label_encoder.fit_transform(x_train[col])
        x_test[col] = label_encoder.transform(x_test[col])

    # Define the transformer pipeline with hyperparameters
    transformer = ColumnTransformer(transformers=[
        ("tnf1", OneHotEncoder(sparse_output=False, drop="first"), [0, 4, 6]),
        ("tnf3", RobustScaler(), [1, 7, 8])
    ], remainder='passthrough')

    # Set up the pipeline with feature selection
    pipeline = Pipeline([
        ('preprocessing', transformer),
        ('feature_selection', SelectKBest(score_func=f_classif, k=5))
    ])

    # Apply preprocessing and feature selection to training data
    X_train_transformed = pipeline.fit_transform(x_train, y_train)

    # Resample the data using SMOTE
    smote = SMOTE(sampling_strategy='minority', k_neighbors=5)
    X_resampled, y_resampled = smote.fit_resample(X_train_transformed, y_train)

    # Define hyperparameters for CatBoostClassifier
    depth = trial.suggest_int('depth', 4, 10)
    learning_rate = trial.suggest_float('learning_rate', 0.01, 0.3, log=True)
    iterations = trial.suggest_int('iterations', 100, 1000)
    l2_leaf_reg = trial.suggest_float('l2_leaf_reg', 1, 10, step=1)
    border_count = trial.suggest_int('border_count', 32, 255)
    scale_pos_weight = trial.suggest_float('scale_pos_weight', 1, 10)

    # Initialize and fit the model
    model = CatBoostClassifier(
        depth=depth,
        learning_rate=learning_rate,
        iterations=iterations,
        l2_leaf_reg=l2_leaf_reg,
        border_count=border_count,
        scale_pos_weight=scale_pos_weight,  # Adjusts for class imbalance
        loss_function='Logloss',  # Suitable for binary classification
        eval_metric='AUC',  # Optimize for area under the ROC curve
        verbose=0,  # Suppress verbose output
        random_state=42
    )
    model.fit(X_resampled, y_resampled)

    # Evaluate the model on the test set
    X_test_transformed = pipeline.transform(x_test)
    y_pred = model.predict(X_test_transformed)
    y_pred_proba = model.predict_proba(X_test_transformed)[:, 1]

    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    # Return ROC-AUC for Optuna to maximize (better for imbalanced datasets)
    return roc_auc

# Set up Optuna study
study = optuna.create_study(direction="maximize")

# Perform optimization with 20 trials
study.optimize(objective, n_trials=10)

# Print the best hyperparameters and the best ROC-AUC found
print(f"Best trial: {study.best_trial.params}")
print(f"Best ROC-AUC: {study.best_value:.4f}")
```
